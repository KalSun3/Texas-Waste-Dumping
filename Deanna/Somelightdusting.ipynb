```python
#looping over all files
for all_file in all_files:
    file_path = os.path.join(file_folder, all_file)

    df = pd.read_csv(file_path)

    df = df.drop(columns=deleting_columns, errors='ignore')

    df = df.rename(columns=renaming_columns)

    cleaned_files = f"{os.path.splitext(all_file)[0]}_cleaned.csv"
    
    new_path = os.path.join(new_output, cleaned_files)

    df.to_csv(new_path, index=False)
```
